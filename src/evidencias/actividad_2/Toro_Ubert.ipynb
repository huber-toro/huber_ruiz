## Ejercicio 2: Suma de todos los elementos en un array 10x10

Aquí creamos un array de tamaño 10x10 con valores aleatorios entre 1 y 9.  
Luego calculamos la suma de todos sus elementos usando `np.sum()`.  
Este ejercicio refuerza el manejo de matrices y operaciones agregadas en NumPy.


```python
# Crear un array 10x10 con valores aleatorios
array2 = np.random.randint(1, 10, (10, 10))
suma = np.sum(array2)

print("Matriz 10x10:\n", array2)
print("Suma de todos los elementos:", suma)
```

## Ejercicio 3: Multiplicación de dos arrays aleatorios

En este ejercicio generamos dos arrays de NumPy de tamaño 5, cada uno con números aleatorios entre 1 y 10.  
Luego realizamos la multiplicación elemento a elemento (`array1 * array2`).  
Este ejercicio refuerza el uso de operaciones vectorizadas en NumPy, que permiten trabajar con arrays de manera eficiente sin necesidad de bucles explícitos.


```python
# Generar dos arrays de tamaño 5 con números aleatorios entre 1 y 10
array1 = np.random.randint(1, 11, 5)
array2 = np.random.randint(1, 11, 5)

# Multiplicación elemento a elemento
resultado = array1 * array2

print("Array 1:", array1)
print("Array 2:", array2)
print("Resultado de la multiplicación:", resultado)
```

## Ejercicio 4: Matriz 4x4 con i+j y su inversa

Aquí construimos una matriz de tamaño 4x4 donde cada elemento se define como la suma de sus índices de fila y columna (`i + j`).  
Luego calculamos la inversa de la matriz usando `np.linalg.inv()`.  
Este ejercicio combina la creación de matrices personalizadas con operaciones algebraicas avanzadas en NumPy.


```python
# Crear una matriz 4x4 aleatoria con valores entre 1 y 10
matriz = np.random.randint(1, 10, (4, 4))

# Calcular la inversa
inversa = np.linalg.inv(matriz)

print("Matriz original:\n", matriz)
print("Matriz inversa:\n", inversa)

```

## Ejercicio 5: Máximo y mínimo en un array de 100 elementos

En este ejercicio generamos un array de 100 números aleatorios entre 0 y 1 usando `np.random.rand()`.  
Luego encontramos el valor máximo y el mínimo con `np.max()` y `np.min()`, respectivamente.  
También identificamos las posiciones (índices) de esos valores con `np.argmax()` y `np.argmin()`.  
Este ejercicio refuerza el uso de funciones de agregación y búsqueda en NumPy.


```python
# Generar un array de 100 elementos aleatorios entre 0 y 1
array5 = np.random.rand(100)

# Calcular máximo y mínimo
max_val = np.max(array5)
min_val = np.min(array5)

# Encontrar índices
indice_max = np.argmax(array5)
indice_min = np.argmin(array5)

print("Array generado:\n", array5)
print("\nValor máximo:", max_val, "en índice", indice_max)
print("Valor mínimo:", min_val, "en índice", indice_min)
```

## Ejercicio 6: Submatriz 2x2 de una matriz 5x5

En este ejercicio generamos una matriz de tamaño 5x5 con números aleatorios entre 1 y 9.  
Luego extraemos una submatriz de tamaño 2x2 que comienza en la segunda fila y segunda columna.  
Este ejercicio refuerza el uso del **indexado en NumPy**, que permite seleccionar porciones específicas de una matriz.


```python
# Generar una matriz 5x5 con valores aleatorios
matriz5x5 = np.random.randint(1, 10, (5, 5))

# Extraer submatriz 2x2 desde la segunda fila y columna
submatriz = matriz5x5[1:3, 1:3]

print("Matriz 5x5:\n", matriz5x5)
print("\nSubmatriz 2x2:\n", submatriz)
```

## Ejercicio 7: Array de ceros y modificación de valores

En este ejercicio creamos un array de tamaño 10 lleno de ceros usando `np.zeros()`.  
Luego modificamos los valores en el rango de índices 3 a 6 para que sean iguales a 5.  
Este ejercicio refuerza el uso del **indexado y asignación en arrays**.


```python
# Crear un array de ceros de tamaño 10
array7 = np.zeros(10)

# Cambiar valores en índices 3 a 6
array7[3:7] = 5

print("Array modificado:", array7)
```

## Ejercicio 8: Invertir filas de una matriz 3x3

En este ejercicio generamos una matriz de tamaño 3x3 con números aleatorios entre 1 y 9.  
Luego invertimos el orden de sus filas utilizando indexado en NumPy (`matriz[::-1]`).  
Este ejercicio refuerza el uso de **slicing** para manipular la estructura de una matriz.


```python
# Generar una matriz 3x3 con valores aleatorios
matriz = np.random.randint(1, 10, (3, 3))

# Invertir el orden de las filas
matriz_invertida = matriz[::-1]

print("Matriz original:\n", matriz)
print("\nMatriz con filas invertidas:\n", matriz_invertida)
```

## Ejercicio 9: Filtrar valores mayores a 0.5 en un array

En este ejercicio generamos un array de tamaño 10 con números aleatorios entre 0 y 1 usando `np.random.rand()`.  
Luego seleccionamos únicamente los valores mayores a 0.5 mediante indexado booleano (`array[array > 0.5]`).  
Este ejercicio refuerza el uso de **condiciones lógicas en NumPy** para filtrar datos.


```python
# Generar un array de 10 elementos aleatorios entre 0 y 1
array9 = np.random.rand(10)

# Filtrar valores mayores a 0.5
mayores_05 = array9[array9 > 0.5]

print("Array generado:\n", array9)
print("\nValores mayores a 0.5:\n", mayores_05)
```

## Ejercicio 10: Gráfico de dispersión con dos arrays aleatorios

En este ejercicio generamos dos arrays de tamaño 100 con números aleatorios entre 0 y 1 usando `np.random.rand()`.  
Luego construimos un gráfico de dispersión para visualizar la relación entre ambos conjuntos de datos.  
Este ejercicio refuerza el uso de **visualización con Matplotlib**, mostrando cómo se distribuyen los puntos en el plano.



```python
import matplotlib.pyplot as plt

# Generar dos arrays aleatorios de tamaño 100
X = np.random.rand(100)
Y = np.random.rand(100)

# Crear gráfico de dispersión
plt.figure()
plt.scatter(X, Y)
plt.title("Gráfico de dispersión")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

## Ejercicio 11: Datos periódicos con ruido

En este ejercicio generamos un conjunto de datos periódicos usando la función `sin(x)` en el rango de -2π a 2π.  
Luego añadimos una pequeña perturbación aleatoria (ruido gaussiano) para simular datos experimentales.  
Este ejercicio refuerza el uso de funciones matemáticas y la simulación de ruido en datos.


```python
# Crear datos periódicos con ruido
x = np.linspace(-2*np.pi, 2*np.pi, 100)
y_ruido = np.sin(x) + np.random.normal(0, 0.1, 100)

print("Primeros 10 valores de x:", x[:10])
print("Primeros 10 valores de y con ruido:", y_ruido[:10])
```

## Ejercicio 12: Gráfico con ruido y curva de referencia

En este ejercicio representamos en un mismo gráfico dos conjuntos de datos:  
- Los datos periódicos con ruido gaussiano (`sin(x) + ruido`) mediante un gráfico de dispersión.  
- La curva de referencia (`sin(x)`) mediante una línea roja.  

Este ejercicio refuerza la comparación visual entre datos experimentales (con ruido) y la función teórica de referencia.



```python
# Crear datos periódicos con ruido
x = np.linspace(-2*np.pi, 2*np.pi, 100)
y_ruido = np.sin(x) + np.random.normal(0, 0.1, 100)

# Representar en un mismo gráfico
plt.figure()
plt.scatter(x, y_ruido, label="sin(x) + ruido")
plt.plot(x, np.sin(x), color="red", label="sin(x)")
plt.legend()
plt.title("Gráfico con ruido y referencia")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

## Ejercicio 13: Mejorar gráfico con etiquetas y título

En este ejercicio tomamos el gráfico del ejercicio anterior y lo mejoramos agregando:  
- Etiqueta para el eje X: **Eje X**  
- Etiqueta para el eje Y: **Eje Y**  
- Título: **Gráfico de Dispersión**  
- Leyendas con notación LaTeX para mayor claridad.  

Este ejercicio refuerza la presentación profesional de gráficos en Matplotlib.


```python
# Crear datos periódicos con ruido
x = np.linspace(-2*np.pi, 2*np.pi, 100)
y_ruido = np.sin(x) + np.random.normal(0, 0.1, 100)

# Gráfico mejorado con etiquetas y título
plt.figure()
plt.scatter(x, y_ruido, label=r"$\sin(x) + ruido$")
plt.plot(x, np.sin(x), color="red", label=r"$\sin(x)$")
plt.legend()
plt.title("Gráfico de Dispersión")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.show()
```
